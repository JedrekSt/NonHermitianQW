<h1>Description</h1>

Hilbert space is composed of three parts:
$$
H = H_{p}\otimes H_{c} \otimes H_{r}
$$
position space, coin space, and rabbit space.

We define standard evoution operators as follows:
$$
S = \sum_{x=0}^{d-1}(|x+1,\rightarrow\rangle \langle x, \rightarrow|  + |x-1,\leftarrow\rangle \langle x, \leftarrow|) \otimes I_{r}
$$
$$
C = \sum_{x=0}^{d-1}\sum_{y=0}^{d-1}|x\rangle \langle x | \otimes \big[(\delta_{x,y}+\delta_{x-1,y})\sigma_{x}+(1-\delta_{x,y}-\delta_{x-1,y})C_{\hat{n}}(\theta)\big] \otimes  |y\rangle \langle y |
$$
if the walker is in the postion $x=y$ or in the postion $x=y+1$, then the coin operator takes form of $\sigma_{x}$. Otherwise one should apply standard rotation operator $C_{\hat{n}}(\theta)=\exp(-i\hat{n}\cdot\vec{\sigma}\theta)$


Krauss operators
$$
K_{0} = \sum_{\hat{y} \in \psi_{0}}|\hat{y} , \leftarrow , \hat{y}\rangle \langle \hat{y} , \rightarrow \hat{y}|
$$
$$
K_{1} = \sqrt{I-K_{0}^{+}K_{0}}
$$
Total evoluton driven by:
$$
\rho(t+1) = SC\big(\sum_{i=0}^{1}K_{i}\rho(t) K_{i}^{+}\big) C^{+} S^{+}
$$

```python
def vis(data,negativity,rabbit_prob):
    fig,axs = plt.subplots(1,3,figsize=(17,6))
    axs[0].pcolormesh(np.real(data),cmap="inferno")
    axs[0].set_xlabel("position")
    axs[0].set_ylabel("steps")
    axs[0].set_title("probability distribution")

    axs[1].set_xlabel("time")
    axs[1].set_ylabel("negativity")
    axs[1].set_title("negativity")

    axs[1].plot(range(len(np.real(negativity))),np.round(np.real(negativity),7),color="red",lw = 1)

    axs[2].set_xlabel("states")
    axs[2].set_ylabel("steps")
    axs[2].set_title("rabbit probability")

    axs[2].pcolormesh(np.abs(rabbit_prob),cmap="inferno")
    plt.show()
```

```python
class simulation_ent(synch_QW):
    def psi0_(self,x0,a,b,y_):
        p0 =[np.eye(1,self.dim,k=x0[i]).flatten() for i in range(len(x0))]
        y_tab = [np.eye(1,self.y_dim,k=i).flatten() for i in range(len(y_))]
        ans = sum(np.kron(np.kron(p0[i],np.array([a,b])),y_tab[i]) for i in range(min(len(y_tab),len(x0))))
        return ans / np.sqrt(sum(np.abs(ans[i])**2 for i in range(len(ans))))
```
