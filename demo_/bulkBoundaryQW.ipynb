```python
import numpy as np
from matplotlib import pyplot as plt
from QW1dim import *
```

```python
dim = 40
coin_dict = {
    "thA" : -np.pi/2,
    "thB" : np.pi/2
}
init_state = {
    "x0" : 10,
    "a" : 1,
    "b" : 1j
}
model = TopologicalPhaseQW(dim,coin_dict,**init_state)
data = model.evolve(100)
```

```python
fig = plt.figure(figsize = (5,5))
ax = fig.add_subplot()

ax.imshow(data,origin="lower",aspect="auto")
ax.set_xlabel("x")
ax.set_ylabel("t")
plt.show()
```

```python
dim = 40
coin_dict = lambda thA,thB : {
    "thA" : thA,
    "thB" : thB
}
init_state = lambda x0,a,b : {
    "x0" : x0,
    "a" : a,
    "b" : b
}

thA_tab = np.arange(-np.pi,np.pi,0.1)
inds_sort = np.array([0, dim-1,dim,2*dim-1])
E = []
states = []
for i in range(len(thA_tab)):
    model = TopologicalPhaseQW(dim,coin_dict(thA_tab[i],np.pi/2))
    En,U = np.linalg.eig(model.U_)
    inds = np.argsort(np.angle(En))
    En , U = np.angle(En[inds]), U[:,inds]
    E.append(En)
    states.append(U[:,inds_sort])
E = np.array(E)

fig = plt.figure(figsize = (10,3))
ax = fig.add_subplot(121)
for i in range(E.shape[1]):
    ax.plot(thA_tab,E[:,i],color = "red",alpha = 0.7,lw = 1)
ax.grid(True)
ax.set_xlabel(r"$\theta_{A}$")
ax.set_ylabel(r"$E$")
ax.set_title(r"$\theta_{B} = \pi/2$")

ax2 = fig.add_subplot(122)
it = len(thA_tab)//2-10
prob = np.abs(states[it][0::2,3])**2 + np.abs(states[it][1::2,3])**2
ax2.bar(range(dim),prob)
ax2.set_xlabel("x")
ax2.set_ylabel("Probability")
ax2.set_title("Edge state")


plt.show()
```
